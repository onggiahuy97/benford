```python
import numpy as np
import matplotlib.pyplot as plt

# Benford probabilities for first digit
def benford_first_digit():
    return np.array([np.log10(1 + 1/d) for d in range(1, 10)])

# Benford probabilities for first two digits
def benford_first_two_digits():
    return np.array([np.log10(1 + 1/d) for d in range(10, 100)])

# Benford probabilities for first three digits
def benford_first_three_digits():
    return np.array([np.log10(1 + 1/d) for d in range(100, 1000)])

# Compute distributions
p1 = benford_first_digit()
p2 = benford_first_two_digits()
p3 = benford_first_three_digits()

# Plot
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# First digit
axs[0].bar(range(1, 10), p1)
axs[0].set_title("Benford's Law - First Digit")
axs[0].set_xlabel("Digit")
axs[0].set_ylabel("Probability")
axs[0].set_xticks(range(1, 10))

# First two digits
axs[1].bar(range(10, 100), p2)
axs[1].set_title("Benford's Law - First Two Digits")
axs[1].set_xlabel("Two-digit number")
axs[1].set_ylabel("Probability")
axs[1].set_xticks(range(10, 100, 10))

# First three digits
axs[2].bar(range(100, 1000), p3)
axs[2].set_title("Benford's Law - First Three Digits")
axs[2].set_xlabel("Three-digit number")
axs[2].set_ylabel("Probability")
axs[2].set_xticks(range(100, 1000, 100))

plt.tight_layout()
plt.show()
```
